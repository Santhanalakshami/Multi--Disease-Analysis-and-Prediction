```python
# Step 1: Import libraries
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (accuracy_score, classification_report, confusion_matrix, 
                             roc_auc_score, roc_curve)


```

```python
#Step 2: Split Features and Target

X = parkinson_df.drop('status',axis=1)
y = parkinson_df['status']
```

```python
#Step 3: Scale Features

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Step 4: Train & Test Split

X_train , X_test , y_train , y_test = train_test_split(X_scaled, y,test_size=0.2,random_state=42,stratify=y)
```

```python
#Step 5a: Gridsearch CV for logistic regression
    
lr = LogisticRegression(max_iter=1000,class_weight='balanced')
param_grid_lr = {
    'C': [0.01, 0.1, 1, 10],
    'penalty': ['l2'],
    'solver': ['lbfgs']
}
grid_lr = GridSearchCV(lr, param_grid_lr, cv=5, scoring='accuracy', n_jobs=-1)
grid_lr.fit(X_train, y_train)
best_lr = grid_lr.best_estimator_
```

```python
#Predictions and Probablities

y_pred_lr = best_lr.predict(X_test)
y_proba_lr = best_lr.predict_proba(X_test)[:,1]
```

```python
#Step 5b: Gridsearch CV for randomforest

rf = RandomForestClassifier(random_state=42)
param_grid_rf = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
grid_rf = GridSearchCV(rf, param_grid_rf, cv=5, scoring='accuracy', n_jobs=-1)
grid_rf.fit(X_train, y_train)
best_rf = grid_rf.best_estimator_
```

```python
#Predictions and Probablities

y_pred_rf = best_rf.predict(X_test)
y_proba_rf = best_rf.predict_proba(X_test)[:,1]
```

```python
# Logistic Regression
print("=== Logistic Regression ===")
print("Best Parameters:", grid_lr.best_params_)
print("Accuracy:", accuracy_score(y_test, y_pred_lr))
print("Classification Report:\n", classification_report(y_test, y_pred_lr))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_lr))
roc_auc_lr = roc_auc_score(y_test, y_proba_lr)
print("ROC-AUC Score:", roc_auc_lr)
```

```python
# Random Forest
print("\n=== Random Forest ===")
print("Best Parameters:", grid_rf.best_params_)
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Classification Report:\n", classification_report(y_test, y_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
roc_auc_rf = roc_auc_score(y_test, y_proba_rf)
print("ROC-AUC Score:", roc_auc_rf)
```

```python
# Step 7: Cross-validation (5-fold)
cv_lr = cross_val_score(best_lr, X_scaled, y, cv=5, scoring='accuracy')
cv_rf = cross_val_score(best_rf, X_scaled, y, cv=5, scoring='accuracy')
print("\nLogistic Regression 5-Fold CV Accuracy:", cv_lr)
print("Mean CV Accuracy:", cv_lr.mean())
print("\nRandom Forest 5-Fold CV Accuracy:", cv_rf)
print("Mean CV Accuracy:", cv_rf.mean())
```

```python
# Step 8: Confusion Matrix Plots

# Logistic Regression
plt.figure(figsize=(6,5))
sns.heatmap(confusion_matrix(y_test, y_pred_lr), annot=True, fmt='d', cmap='RdBu', cbar=False)
plt.title("Logistic Regression Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Random Forest
plt.figure(figsize=(6,5))
sns.heatmap(confusion_matrix(y_test, y_pred_rf), annot=True, fmt='d', cmap='Spectral', cbar=False)
plt.title("Random Forest Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
